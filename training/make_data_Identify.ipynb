```python
import pandas as pd
import shutil
import os

csv_file_path = '/mnt/driver/STT_data/STT_dataset/64.Metadata_t5_64_63_..._57RDC/64.Metadata_t6__exits.csv'
output_folder = '/mnt/driver/Identify_data/vi'

df = pd.read_csv(csv_file_path)
df = df.sample(frac=1).reset_index(drop=True)
df = df.sample(frac=1).reset_index(drop=True)
df = df.sample(frac=1).reset_index(drop=True)
df = df.sample(frac=1).reset_index(drop=True)
df = df.sample(frac=0.2, random_state=42).reset_index(drop=True)
for index, row in df.iterrows():
    audio_file_path = row['file']
    duration = row['duration']
    new_filename = f"1_{duration}_{index}.wav"
    output_file_path = os.path.join(output_folder, new_filename)
    shutil.copy(audio_file_path, output_file_path)
    print(f"Copy {audio_file_path} to {output_file_path}")

print("Copy hoàn thành!")
```
