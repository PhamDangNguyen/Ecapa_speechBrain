```python
import scipy.io.wavfile as wavfile
def compute_duration(filename):
    sampling_rate, data = wavfile.read(filename)
    duration = len(data) / sampling_rate
    return duration
```

```python
import numpy as np
import scipy.io.wavfile as wavfile
import subprocess
import os  # For optional file deletion

sampling_rate = 16000
count = 0
for index in range(len(ds["train"])):
    data = ds["train"][index]["audio"]["array"]
    temp_wav_file = f"/home/pdnguyen/Identification_ASR_practice/valid_data_Identify/44khz/{index}.wav"
    wavfile.write(temp_wav_file, sampling_rate, (data * 32767).astype(np.int16))
    duration = compute_duration(temp_wav_file)
    output_16k = f"/home/pdnguyen/Identification_ASR_practice/valid_data_Identify/vi/0_{duration}_{count}.wav"
# Construct FFmpeg command string
    ffmpeg_cmd = f"ffmpeg -i {temp_wav_file} -ar 16000 -ac 1 -acodec pcm_s16le {output_16k}"
    subprocess.run(ffmpeg_cmd.split())  # Split the command string into arguments
    count = count + 1
    if count == 200:
        break
print("-----------------------------------------",count,"-----------------------------------------")
```
